```python
import pandas as pd
```

```python
euro_leagues = pd.read_csv('D:/Manipal Data Science/First Term/NumPy/goal_stats_euro_leagues_2012-13.csv')
euro_leagues
```

```python
euro_goals = euro_leagues.set_index('Month')
euro_goals
```

```python
euro_grpby_year = euro_goals.groupby(lambda Month: Month.split('/')[2])
euro_grpby_year.groups
```

```python
for  name,group in euro_grpby_year:
    print(name)
    print(group)
```

```python
euro_grpby_month = euro_goals.groupby(level = 0)
euro_grpby_month
```

```python
euro_grpby_month.groups
```

```python
for i,j in euro_grpby_month:
    print(i,j)
    print("\n")
```
