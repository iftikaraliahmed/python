```python
gap_data = pd.read_csv("./DATA/gapminder.csv")
gap_data.head()
```

```python
gap_data_new=gap_data.set_index(['Year','region','Country'])
gap_data_new.head()
```

```python
def spredd(x):
    result = x.max() - x.min()
    return result
```

```python
by_year_region = gap_data.groupby(['Year','region']).agg({"population":sum,"child_mortality":np.mean,"gdp":spredd})
by_year_region.head()
```
