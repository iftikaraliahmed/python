```python
import pandas as pd
tech_stock=pd.read_csv('D:/Manipal Data Science/First Term/NumPy/tech_stockprices.csv')
tech_stock
```

```python
tech_stock.set_index('Symbol')
```

```python
##Slice the first 4 rows and Closing price, EPS cols

techslice1 = tech_stock.iloc[:4,1:3]
techslice1
```

```python
##Slice starting from row 2,row 3 and only the col P/E

techslice2 = tech_stock.iloc[2:4, -2:-1]
techslice2
```

```python
##Slice from row 1 to 5 and only the col Market Cap(B)

techslice3 = tech_stock.iloc[1:,-1:]
techslice3
```

```python
pd.concat([techslice1,techslice2,techslice3],axis=1)
#deafult join method is outer join
```

```python
import numpy as np
np.random.seed(123)
norm_dist = pd.DataFrame(np.random.randn(1,4))
norm_dist
```

```python
binom_dist = pd.DataFrame(np.random.binomial(100,0.5,(1,4)))
binom_dist
```

```python
poisson_dist = pd.DataFrame(np.random.poisson(100,(1,4)))
poisson_dist
```

```python
dist = [norm_dist,binom_dist,poisson_dist]
dist
```

```python
df2 = pd.concat(dist,keys=['Normal','Binomial','Poisson'])
df2
```

```python
#Slice the first 2 rows & first 3 columns

slice1=df2.ix[:2,:3]
slice1
```

```python
#slice the remaining row
slice2 = df2[2:] 
slice2

#slice2=df2.loc['Poisson']     
#slice2
```

```python
con = pd.concat([slice1,slice2],axis=0)
con
```
