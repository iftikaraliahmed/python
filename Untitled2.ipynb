```python
import pandas as pd
us_data=pd.read_csv('D:/Manipal Data Science/First Term/NumPy/us_index_data.csv')
us_data
```

```python
us_data=us_data.set_index('TradingDate')
us_data
```

```python
#Slice the first 2 rows and 3 columns

slice1 = us_data.iloc[:2,:3]
slice1
```

```python
#Slice the first 2 rows & col 0,3,4

slice2 = us_data.iloc[:2,[0,2,3]]
slice2
```

```python
##Slice rows 1 & 2 and first 3 cols

slice3 = us_data.iloc[1:3,:3]
slice3
```

```python
#merging rows

pd.merge(slice1,slice2)
```

```python
#slice inner

pd.merge(slice2,slice3,how='inner')
```

```python
#slice outer

pd.merge(slice2,slice3,how='outer')
```

```python
pd.merge(slice2,slice3,how='left')
```

```python
pd.merge(slice2,slice3,how='right')
```
